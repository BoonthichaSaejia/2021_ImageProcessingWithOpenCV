# **0 Import Library and Data**

```
import cv2
from google.colab.patches import cv2_imshow
import numpy as np
```

```
image = cv2.imread('/content/amulet.jpg')
cv2_imshow(image)
```

# **1 Blur Image**

```
gblur1 = cv2.GaussianBlur(image,(3,3),0)
gblur2 = cv2.GaussianBlur(image,(5,5),0)
gblur3 = cv2.GaussianBlur(image,(7,7),0)
gblur4 = cv2.GaussianBlur(image,(9,9),0)
gblur5 = cv2.GaussianBlur(image,(11,11),0)
gblur6 = cv2.GaussianBlur(image,(13,13),0)
gblur7 = cv2.GaussianBlur(image,(21,21),0)
```

```
cv2_imshow(gblur1)
cv2_imshow(gblur7)
```

```
gray1 = cv2.cvtColor(gblur1,cv2.COLOR_BGR2GRAY)
gray2 = cv2.cvtColor(gblur2,cv2.COLOR_BGR2GRAY)
gray3 = cv2.cvtColor(gblur3,cv2.COLOR_BGR2GRAY)
gray4 = cv2.cvtColor(gblur4,cv2.COLOR_BGR2GRAY)
gray5 = cv2.cvtColor(gblur5,cv2.COLOR_BGR2GRAY)
gray6 = cv2.cvtColor(gblur6,cv2.COLOR_BGR2GRAY)
gray7 = cv2.cvtColor(gblur7,cv2.COLOR_BGR2GRAY)
```

# **2 Find threshold**

```
(T,thresh1) = cv2.threshold(gray1,70,255,cv2.THRESH_BINARY)
(T,thresh2) = cv2.threshold(gray2,70,255,cv2.THRESH_BINARY)
(T,thresh3) = cv2.threshold(gray3,70,255,cv2.THRESH_BINARY)
(T,thresh4) = cv2.threshold(gray4,70,255,cv2.THRESH_BINARY)
(T,thresh5) = cv2.threshold(gray5,70,255,cv2.THRESH_BINARY)
(T,thresh6) = cv2.threshold(gray6,70,255,cv2.THRESH_BINARY)
(T,thresh7) = cv2.threshold(gray7,70,255,cv2.THRESH_BINARY)
```

```
cv2_imshow(thresh1)
cv2_imshow(thresh2)
cv2_imshow(thresh3)
cv2_imshow(thresh4)
cv2_imshow(thresh5)
cv2_imshow(thresh6)
cv2_imshow(thresh7)
```

# **3 Bitwise_and 2 image**

```
onlyAmu = cv2.bitwise_and(gblur7,gblur7,mask=thresh7)
```

```
cv2_imshow(onlyAmu)
```

# **4 Find Edge**

```
canny = cv2.Canny(onlyAmu,15,240)
cv2_imshow(canny)
```

# **5 Contour the external edge**

```
(cnts,_) = cv2.findContours(canny,cv2.RETR_EXTERNAL,cv2.CHAIN_APPROX_SIMPLE)
```

```
lenAmu = len(cnts)
```

```
print(f"The number of amulet in the image is : {lenAmu} ")
```

```
amulets = image.copy()
```

```
cv2.drawContours(amulets,cnts,-1,(0,255,0),1)
cv2_imshow(amulets)
```
